```python
%matplotlib inline
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis,y_axis, marker = "d", color = "green", label = "Boston", linewidth = 2)
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.ylim(0,45)
plt.grid()
plt.legend()
```

```python
plt.barh(x_axis,y_axis, color = "pink", label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.xlim(0,45)
plt.gca().invert_yaxis()
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis,color = "skyblue", label = "Chicago")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
ax.legend()
```

```python
plt.scatter(y_axis,x_axis,c="red", label= "Chicago")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,45)
plt.gca().invert_yaxis()
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis,y_axis,c="red", label= "Chicago", s=y_axis_larger)
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.ylim(0,45)
plt.legend()
```

```python
plt.scatter(y_axis,x_axis,c="skyblue", label= "Chicago",s=[i * 5 for i in y_axis], alpha = 0.8, linewidth = 2, 
            edgecolor = "black")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        explode=explode_values, 
        labels=x_axis, 
        autopct='%.1f%%', 
        colors = colors)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
explode_list = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, labels = x_axis, autopct = "%.1f%%",
      explode = explode_list,
      startangle=90,
      counterclock=False,
      colors = colors)
```

```python
stdev =statistics.stdev(y_axis)
stdev
```

```python
plt.bar(x_axis,y_axis,yerr=stdev, capsize = 3)
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
